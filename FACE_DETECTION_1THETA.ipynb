```python
import tensorflow as tf
import os
import cv2
import numpy as np
from tensorflow.keras.preprocessing import image
os.chdir(r"C:\Users\Hpane\OneDrive\Desktop\face detection")
cnn=tf.keras.models.load_model(r"C:\Users\Hpane\OneDrive\Desktop\face detectioncnn.keras")
cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python


#First step-> Extract out the frame of face 
cap=cv2.VideoCapture(0)

while cap.isOpened():
    b,img=cap.read()
    faces=cascade.detectMultiScale(img,scaleFactor=1.10,minNeighbors=4)
    #This function will tell the position of face in frame
    #detectmultiscale will move over surface of image and match the haarcacade features and return position
    #where it finds the face
    #scalefactor-> How much to compress and decompress image for detection
    #minneighbors-> When window is convolve over surface if it find face 4 times continously
    #only then that thing will be considered as face otherwise it may be noise
    for x,y,w,h in faces:
        face=img[y:y+h,x:x+w]#Slice out face from image
        cv2.imwrite("face.jpg",face)#Saving that face
        face=image.load_img("face.jpg",target_size=(150,150))
        #Now image is in the keras format-> we have to convert it into numpy array
        face=image.img_to_array(face)#Convert into numpy array
        face=np.expand_dims(face,axis=0)
        #This fxn is use to convert data into higher dimension
        #axis=0 means to expand first dimension
        #(150,150,3)->(1,150,150,3)
        
        ans=cnn.predict(face)
        if ans>0.5:
            cv2.rectangle(img,(x//2,y//2),(x+w,y+h),(0,0,255),3)
            cv2.putText(img,'Without Mask',(x,y),2,4,(255,255,255))
            print("Without mask")
        else:
            cv2.rectangle(img,(x//2,y//2),(x+w,y+h),(0,255,0),3)
            cv2.putText(img,'With mask',(x,y),2,4,(255,255,255))
            print("With mask")
    cv2.imshow("frame",img)
    if cv2.waitKey(1)==97:
        break
cap.release()
cv2.destroyAllWindows()
#cnn will consider first dimension of data as total samples
#So we have to send data in 4D for prediction
```
